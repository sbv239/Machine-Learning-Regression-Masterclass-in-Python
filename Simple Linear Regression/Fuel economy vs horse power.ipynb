```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fuel_ec = pd.read_csv('FuelEconomy.csv')
fuel_ec.describe()
```

```python
fuel_ec.info()
```

```python
sns.jointplot(y = 'Fuel Economy (MPG)',
              x = 'Horse Power',
              data = fuel_ec,
              color = 'blue')
```

```python
sns.pairplot(fuel_ec)
```

```python
sns.lmplot(y = 'Fuel Economy (MPG)',
              x = 'Horse Power',
              data = fuel_ec)
```

```python
X = fuel_ec[['Horse Power']]
y = fuel_ec['Fuel Economy (MPG)']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
lr = LinearRegression(fit_intercept=True)
lr.fit(X_train, y_train)
```

```python
print('Linear Model Coeff (slope): ', lr.coef_)
print('Linear Model Coeff (bias): ', lr.intercept_)
```

```python
y_predict = lr.predict(X_test)
```

```python
plt.scatter(X_train, y_train, color = 'gray')
plt.plot(X_train, lr.predict(X_train), color = 'blue')
plt.ylabel('Fuel economy, MPG')
plt.xlabel('Horse Power, HP')
plt.title('Horse power vs. Fuel economy (train data)')
```

```python
plt.scatter(X_test, y_test, color = 'gray')
plt.plot(X_test, y_predict, color = 'red')
plt.ylabel('Fuel economy, MPG')
plt.xlabel('Horse Power, HP')
plt.title('Horse power vs. Fuel economy (test data)')
```
