```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
sales = pd.read_csv('EconomiesOfScale.csv')
sales.head()
```

```python
sns.scatterplot(x='Number of Units', y = 'Manufacturing Cost', data = sales)
```

```python
X_train = sales[['Number of Units']]
y_train = sales[['Manufacturing Cost']]
```

### Polynomial model

```python
poly = PolynomialFeatures(degree = 4)

X_colunmns = poly.fit_transform(X_train)

model = LinearRegression(fit_intercept = True)
model.fit(X_colunmns, y_train)
y_predict = model.predict(X_colunmns)

print('Linear model score:', model.score(X_colunmns, y_train))
```

```python
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, y_predict, color = 'blue')
plt.xlabel('Number of Units')
plt.ylabel('Manufacturing Cost')
plt.title('Unit Cost vs. Number of Units [in Millions](Training dataset, poly)')
```

### Linear model

```python
model = LinearRegression(fit_intercept = True)
model.fit(X_train, y_train)
y_predict = model.predict(X_train)

print('Linear model score:', model.score(X_train, y_train))
```

```python
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, y_predict, color = 'blue')
plt.xlabel('Number of Units')
plt.ylabel('Manufacturing Cost')
plt.title('Unit Cost vs. Number of Units [in Millions](Training dataset, linear)')
```
